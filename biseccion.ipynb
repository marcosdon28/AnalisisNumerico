Metodo de la biseccion para encontrar raices


```python
from math import *
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd


#ingreso de parametros
print("Ingrese la funcion")
f = input()
f = eval("lambda x: " + f)
print("Ingrese a del intervalo")
a = float(input())
print("Ingrese b del intervalo")
b = float(input())
print("Ingrese el error de tolerancia")
tol = float(input())
print("Ingrese el número máximo de iteraciones")
max_iter = int(input())

"""
    f : funcion
    a : limite inferior
    b : limite superior
    tol : toleracia o error
    max_iter : numero maximo de iteraciones por si el metodo no converge o converge muy lento

"""

def biseccion(f, a, b, tol, max_iter):
    iter = 0
    x = (a + b) / 2

    #creacion de dataframe para mostrar los datos
    df = pd.DataFrame(columns=['a', 'b', 'c', 'f(a)', 'f(c)', 'f(a)*f(c)', '|E|'])

    while abs(f(x)) > tol:
        if f(a) * f(x) > 0:
            aporc = '>0'
        else:
            aporc = '<0'

        if df.empty:
            df.loc[len(df)] = [a, b, x, f(a), f(x), aporc, '-']

        if f(a) * f(x) < 0:
            b = x
        else:
            a = x
        x = (a + b) / 2

        #agregar datos al dataframe
        df.loc[len(df)] = [a, b, x, f(a), f(x), aporc, abs(f(x))]
        iter += 1

        if iter >= max_iter:
            print("No se pudo encontrar una raíz dentro de los límites dados después de", max_iter, "iteraciones.")
            return None
    
    #mostrar dataframe
    print(df)
    print("Se encontró una raíz en", iter, "iteraciones.")
    print("La raíz es:", x)
    return 

biseccion(f, a, b, tol, max_iter)


```
